### Scrape For CSV Data

```python
from bs4 import BeautifulSoup as bs
import pandas as pd

# Thanks to the website at GitHub for the data!
data_url = "https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_daily_reports_us/"
data_dfs = []
data_count = 0

for month in range(3, 8):
    for x in range(31):
        if x < 9:
            csv_name = f"0{month + 1}-0{x + 1}-2020.csv"
        else:
            csv_name = f"0{month + 1}-{x + 1}-2020.csv"

        try:
            data_tables = pd.read_html(data_url + csv_name)
            data_dfs.append(data_tables[0])
            data_count += 1
        except:
            print(f"404-ish: The csv '{csv_name}' was not found")

data_dfs[-1].head()
```

### Clean the Data

```python
print(data_count)
for i in range(data_count):
    try:
        data_dfs[i] = data_dfs[i].drop(columns='Unnamed: 0')
    except KeyError:
        print(f"The column 'Unnamed: 0' is not in dataframe {i}")
data_dfs[20].head()
```

### Now Put it all in one DataFrame

```python
main_df = pd.DataFrame()

for df in data_dfs:
    main_df = main_df.append(df)

main_df.to_csv('Resources/US_Data.csv')
# main_df.columns
# main_df.head(5)
```

```python
main_df.groupby(['Province_State', 'Last_Update'])['Confirmed', 'Deaths', 'Recovered'].max()
```

### Machine Learning Testing

```python
MN_df = main_df.loc[
    main_df['Province_State'] == 'Minnesota',
    ['Last_Update', 'Confirmed', 'Deaths', 'Recovered', 'Active']
]
MN_df = MN_df.dropna()

# MN_df = MN_df.append(main_df.loc[
#     main_df['Province_State'] == 'Minnesota',
#         ['Last_Update', 'Confirmed', 'Deaths', 'Recovered', 'Active']
# ])
MN_df.head(5)
```

```python
import matplotlib.pyplot as plt

X_data = range(MN_df.groupby('Last_Update').max().count().max())
Y_data = []
for i in range(MN_df.groupby('Last_Update').max().Confirmed.count()):
    Y_data.append(MN_df.groupby('Last_Update').max().Confirmed[i])
    
plt.scatter(X_data, Y_data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
import numpy as np

x_feed = MN_df[['Confirmed', 'Deaths']]
y_feed = MN_df['Last_Update'].values.reshape(-1, 1)
# y = foam["time"].values.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(
    x_feed, y_feed
)

model = SVC(kernel='linear')
model.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))

print('Accuracy: %.3f' % model.score(X_test, y_test))
```
