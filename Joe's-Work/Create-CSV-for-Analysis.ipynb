### Scrape For CSV Data

```python
import pandas as pd

# Thanks to the website at GitHub for the data!
data_url = "https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_daily_reports_us/"
data_dfs = []
data_count = 0

for month in range(3, 8):
    for x in range(31):
        if x < 9:
            csv_name = f"0{month + 1}-0{x + 1}-2020.csv"
        else:
            csv_name = f"0{month + 1}-{x + 1}-2020.csv"

        try:
            data_tables = pd.read_html(data_url + csv_name)
            data_dfs.append(data_tables[0])
            data_count += 1
        except:
            print(f"The csv '{csv_name}' was not found")

data_dfs[-1].head()
```

### Clean the Data

```python
print(data_count)
for i in range(data_count):
    try:
        data_dfs[i] = data_dfs[i].drop(columns='Unnamed: 0')
    except KeyError:
        print(f"The column 'Unnamed: 0' is not in dataframe {i}")
data_dfs[20].head()
```

### Now Put it all in one DataFrame (and Save to CSV)

```python
main_df = pd.DataFrame()

for df in data_dfs:
    main_df = main_df.append(df)

main_df.to_csv('Resources/US_Data.csv')
main_df.head(5)
```
